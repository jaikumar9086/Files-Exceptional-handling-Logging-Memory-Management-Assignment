Q1.  What is the difference between interpreted and compiled languages?

Ans.  An interpreted language is one in which the source code is translated and executed line by line by an interpreter at runtime. Because the translation happens during execution, interpreted languages are generally slower. Errors are detected one at a time as the program runs, which can make debugging more incremental. However, they are usually more portable, since the same code can run on different systems as long as the appropriate interpreter is available. Examples include Python, JavaScript, Ruby, and PHP.

In contrast, a compiled language is one in which the entire source code is translated into machine code (binary) by a compiler before the program is executed. This pre-compilation process usually makes compiled languages faster, since the program is already in machine-readable form when it runs. Errors are typically detected all at once during the compilation stage, before execution begins. Compiled programs are generally less portable because the generated machine code is often specific to a particular platform or operating system. Examples include C, C++, Rust, and Go

Q9.  What is the difference between multithreading and multiprocessing?

Ans.  Both multithreading and multiprocessing are ways to make a program do multiple tasks at the same time (parallel execution) —
but they work in different ways.
________________________________________
🔹  Multithreading

•	Uses multiple threads (smaller units) within the same process.

•	All threads share the same memory.

•	Best for I/O-bound tasks (like reading files, downloading, waiting for input).

•	Limited by Python’s GIL (Global Interpreter Lock) — only one thread runs Python code at a time.

🔹  Multiprocessing

•	Uses multiple processes, each with its own Python interpreter and memory space.

•	Runs true parallel execution on multiple CPU cores.

•	Best for CPU-bound tasks (like calculations, data processing).

•	No GIL issue — each process runs independently.

Q15.  How does Python's garbage collection system work?

Ans.  Garbage collection (GC) is Python’s way of automatically managing memory — allocating memory to objects when needed and freeing it when objects are no longer in use.

Python uses reference counting and a cyclic garbage collector to clean up memory.
________________________________________
1. Reference Counting

•	Every object in Python keeps track of how many references point to it.

•	When the reference count drops to zero, the object is no longer accessible and is deleted automatically.

2. Cyclic Garbage Collector

•	Sometimes objects reference each other, forming a cycle.

•	Reference counting alone cannot detect cycles.

Q23.  What are the challenges associated with memory management in Python?

Ans.  1. Memory Leaks

•	Occur when objects are no longer needed but are still referenced, so memory is not freed.

•	Can happen with global variables, caches, or circular references not handled properly.
________________________________________
2. Circular References

•	Python’s reference counting cannot handle cycles (objects referencing each other).

•	The garbage collector detects cycles, but complex cycles may not be collected immediately, causing temporary memory usage spikes.
________________________________________
3. Large Object Management

•	Large objects (big lists, dictionaries, or images) consume significant memory.

•	Inefficient handling may lead to high memory usage or slow performance.
________________________________________
4. High Memory Overhead

•	Python objects have extra memory overhead due to dynamic typing and garbage collection metadata.

•	Many small objects can use more memory than equivalent C structures.
________________________________________
5. Delayed Garbage Collection

•	The garbage collector runs periodically, not immediately.

•	If memory-intensive operations happen rapidly, Python may temporarily use a lot of memory before cleanup occurs.
________________________________________
6. Resource Cleanup

•	If objects hold resources like file handles, network connections, or database cursors, improper deletion may lead to resource leaks, even if memory is freed.

```python
#Q4.  Write a Python script that reads from one file and writes its content to another file?
#Ans.
source_file = "example.txt" # Assuming this file exists from Q1/Q2
destination_file = "new_example.txt" # New file to write to

try:
  with open(source_file, "r") as source:
    with open(destination_file, "w") as destination:
      for line in source:
        destination.write(line)
      print(f"Content successfully copied from {source_file} to {destination_file}.")

except FileNotFoundError:
  print(f"Error: The file '{source_file}' was not found.")
except Exception as e:
  print(f"An unexpected error occurred: {e}")
```

```python
#Q18.  How would you implement a basic logging setup that logs to a file with rotation after 1MB?
#Ans.
import logging
from logging.handlers import RotatingFileHandler

# Create a logger
logger = logging.getLogger("my_logger")
logger.setLevel(logging.INFO)
handler = RotatingFileHandler("my_log.log", maxBytes=1000000, backupCount=5)
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)
logger.addHandler(handler)
```

```python
#Q21. Write a Python program that reads a file and prints the number of occurrences of a specific word?
#Ans.
filename = "sample.txt"
word_to_count = "python"
count = 0


try:
  with open(filename, "r") as file:
    for line_content in file: # Iterate through the file object to get each line
      words = line_content.lower().split() # Process each line
      count += words.count(word_to_count)
  print(f"The word '{word_to_count}' appears {count} times.") # Closing the f-string
except FileNotFoundError:
  print(f"Error: The file '{filename}' was not found.")
except Exception as e:
  print(f"An unexpected error occurred: {e}")
```
